```python
import pandas as pd
import numpy as np
```

```python
df_sem1 = pd.read_csv(r'.\dataset\sem1.csv')
df_sem2 = pd.read_csv(r'.\dataset\sem2.csv')
df_girls = pd.read_csv(r'.\dataset\girls.csv')
```

```python
# Make all dataset structures same
df_sem1 = df_sem1.drop('Roll Number' , axis = 'columns')
df_sem2 = df_sem2.drop('Roll Number' , axis = 'columns')
```

```python
#merge all datasets
df = pd.concat([df_sem1, df_sem2, df_girls], axis=0)
df.shape
```

```python
df.columns
```

```python
#look for null values
df.isna().sum()
```

```python
#remove all empty data
df = df.dropna()
```

```python
res_cols = df.columns[:-2]
res_cols
```

```python
#look for null values
df.isna().sum()
df.shape
```

```python
#convert strings to lowercase and removing unnecessary spaces
df['type'] = df['type'].apply(lambda x : x.lower())
df['type'] = df['type'].apply(lambda x : x.strip())
df['locality'] = df['locality'].apply(lambda x : x.lower())
df['locality'] = df['locality'].apply(lambda x : x.strip())
```

```python
#Remove unnecessary attributes
df = df.drop('Name' , axis = 'columns')
X = df.drop(['rent' , 'deposit'] , axis = 'columns')
y = df['rent']
```

```python
X.shape
```

```python
len(y)
```

```python
X['type'].unique()
```

```python
#Encoding String attribtes using get_dummies
X = pd.concat([X , pd.get_dummies(X['type'], dtype = 'int')] , axis = 1)
X = pd.concat([X , pd.get_dummies(X['locality'] , dtype = 'int')] , axis = 1)
X = X.drop('type' , axis = 1)
X = X.drop('locality' , axis = 1)
```

```python
#Lone attribute cant define rent
#Plotting scatter plot for hostels 
#commodity_score vs rent
x1 = []
y1 = []

for index, row in df.iterrows():
    if row[4] != 'hostel':
        continue
    x1.append(row[0]) #commodity score
    y1.append(row[-2])  #rent   

from matplotlib import pyplot as plt
plt.scatter(x1, y1)
plt.xlabel('score')
plt.ylabel('rent')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LnR
from sklearn.linear_model import Lasso as L1
from sklearn.linear_model import Ridge as L2
from sklearn.ensemble import RandomForestRegressor as RFG
from xgboost import XGBRegressor as XGB
from sklearn.ensemble import GradientBoostingRegressor as GBR

models = [LnR(), L1(), L2(), RFG(), XGB(), GBR()]
```

```python
scores = []
from sklearn.model_selection import cross_val_score
for model in models:
    score = cross_val_score(model , X , y , cv = 5)
    scores.append(score)
```

```python
for i in range(len(scores)):
    scores[i] = scores[i].mean()
scores
```

```python
scores
```

```python
idx = scores.index(max(scores))
```

```python
model = models[idx]
```

```python
model.fit(X , y)
```

```python
def get_predicted_rent(com_sc , room_sv_sc , mem , size, type_ , room_sc , att_wr , loc , elec , wifi , sec):
    x = np.zeros(18)
    x[0] = com_sc
    x[1] = room_sv_sc
    x[2] = mem
    x[3] = size
    x[4] = room_sc
    x[5] = att_wr
    x[6] = elec
    x[7] = wifi
    x[8] = sec
    
    cols = list(X.columns)
    if loc in cols:
        idx = cols.index(loc)
        x[idx] = 1
        
    if type_ in cols:
        idx = cols.index(type_)
        x[idx] = 1
    return model.predict([x])
```
